```python
#CALCULO DE COORDENADAS ROJA Y AZUL PARA MOBIUS 3D

roja90m3d=[]
rojam3d=[]
azul180m3d=[]
azulm3d=[]

for i in np.arange(0,len(roja90)):
    rojam3d=[]
    for k in np.arange(0,len(roja90[i])//2):
        z1 = roja90[i][2*k]
        if cmath.phase(z1)>0:
            z=z1
            z = (z**2)/(abs(z))
            if np.angle(z)>=0:
                teta = np.angle(z)
            else:
                teta= 2*np.pi+cmath.phase(z)
            v = cmath.rect(1, -teta/2 + np.pi/2)
            w = (3-abs(z))*v
            z = cmath.rect(3,teta)+cmath.rect(w.real, teta)
            zh = complex(w.imag,0)
            rojam3d = np.append(rojam3d, [z,zh])
        if cmath.phase(z1) <= 0:
            z=z1
            z = z*cmath.rect(1,np.pi)
            z = (z**2)/(abs(z))
            if np.angle(z)>=0:
                teta = np.angle(z)
            else:
                teta= 2*np.pi+cmath.phase(z)
            v = cmath.rect(1, -teta/2 + 3*np.pi/2)
            w = (3-abs(z))*v            
            z = cmath.rect(3,teta)+cmath.rect(w.real,teta)
            zh = complex(w.imag,0)
            rojam3d = np.append(rojam3d, [z, zh])
            
    print(i, end=' &  ')
    roja90m3d = roja90m3d+[rojam3d]

for i in np.arange(0,len(azul180)):
    azulm3d=[]
    for k in np.arange(0,len(azul180[i])//2):
        z1 = azul180[i][2*k]
        if cmath.phase(z1)>0:
            z=z1
            z = (z**2)/(abs(z))
            if np.angle(z)>=0:
                teta = np.angle(z)
            else:
                teta= 2*np.pi+cmath.phase(z)
            v = cmath.rect(1, -teta/2 + np.pi/2)
            w = (3-abs(z))*v
            z = cmath.rect(3,teta)+cmath.rect(w.real,teta)
            zh = complex(w.imag,0)
            azulm3d = np.append(azulm3d, [z,zh])
        if cmath.phase(z1) <= 0:
            z=z1
            z = z*cmath.rect(1,np.pi)
            z = (z**2)/(abs(z))
            if np.angle(z)>=0:
                teta = np.angle(z)
            else:
                teta= 2*np.pi+cmath.phase(z)
            v = cmath.rect(1, -teta/2 + 3*np.pi/2)
            w = (3-abs(z))*v            
            z = cmath.rect(3,teta)+cmath.rect(w.real,teta)
            zh = complex(w.imag,0)
            azulm3d = np.append(azulm3d, [z, zh])

    print(i+90, end=' &  ')
    azul180m3d = azul180m3d+[azulm3d]
```
